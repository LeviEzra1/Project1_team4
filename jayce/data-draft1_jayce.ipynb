```python
#Import dependants
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from pathlib import Path
```

```python
CSV1Kpath = Path("Kaggle1DS.csv")
CSV2Kpath = Path("Kaggle2DS.csv")
CSV3Xpath = Path("SHD.csv")
CSV1KR = pd.read_csv(CSV1Kpath)
CSV2KR = pd.read_csv(CSV2Kpath)
CSV3XR = pd.read_csv(CSV3Xpath)

Firstm = pd.merge(CSV1KR, CSV2KR, on="school")
Data = pd.merge(Firstm, CSV3XR, on= "school")

# output DataFrame as CSV to view easier
output_path = "Outputs/merged_df.csv"
Data.to_csv(output_path)
```
